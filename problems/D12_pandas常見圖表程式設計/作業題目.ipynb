```python
import pandas as pd
import numpy as np
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?

boston_data = pd.read_csv('boston.csv')
boston_data.boxplot()
```

```python
#欄位'TAX'和欄位'B'的中位數在300~400之間

print(np.median(boston_data['TAX']))
print(np.median(boston_data['B']))
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
#'NOX'越高'DIS'越低

boston_data.plot.scatter(x = 'NOX', y = 'DIS')
```
